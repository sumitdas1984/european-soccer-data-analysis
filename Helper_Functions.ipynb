```
import json
```

```
def dfToJson(df):
  """
  Dataframe as JSON.
  """
  return (
   df
    .toJSON() # python api only
    .map(lambda r : json.loads(r))
  ).collect()

def jsonColList(jlist, col, default_val, limit=10000):
  """
  Convenience function for column list comprehension.
  """
  result=[]
  c = 0
  for r in jlist:
    try:
        result.append(r[col])
    except Exception as e:
        result.append(default_val)
        pass
    
    c += 1
    if c >= limit:
      return result   
  return result

print("...loaded")
```
